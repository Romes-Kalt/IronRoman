## Simple Rock, Paper, Scissors Game

```python
import random

gestures = ["rock", "paper", "scissors"]
user_points = 0
computer_points = 0
game_counter = 1   # set to one, will only be increased when there is NO TIE

print("Let's play rock, paper and scissors!\n")


while True:
    answer = input("How many rounds to win? ")
    try:
        num_of_wins = int(answer)
    except ValueError:
        print("Please type in a number")
    else:
        round_limit = num_of_wins * 2 - 1
        print("\n\n")
        break
        
while True:
    print(f"Round: {game_counter} of {round_limit}, Player {user_points} : {computer_points} Computer\n")
    user_gesture = "Null"
    while user_gesture.lower() not in ["r", "s", "p", "rock", "paper", "scissors"]:
        user_gesture = input("Enter a choice (r)ock, (p)aper, (s)cissors: ")
        if user_gesture[0].lower() in ["r", "s", "p"]:
            break
        print("Please enter a correct choice.")
    if user_gesture == "r":
        user_gesture = "rock"
    elif user_gesture == "p":
        user_gesture = "paper"
    else:
        user_gesture = "scissors"

    comp_gesture = random.choice(gestures)
    print(f"\nYou chose {user_gesture}, computer chose {comp_gesture}.\n")

    if user_gesture == comp_gesture:
        print(f"Both players selected {user_gesture}. It's a tie!")
    elif user_gesture == "rock":
        if comp_gesture == "scissors":
            print("Rock smashes scissors! You win!")
            user_points += 1
            if user_points == num_of_wins:
                break
            game_counter += 1
        else:
            print("Paper covers rock! You lose.")
            computer_points += 1
            if computer_points == num_of_wins:
                break
            game_counter += 1
    elif user_gesture == "paper":
        if comp_gesture == "rock":
            print("Paper covers rock! You win!")
            user_points += 1
            if user_points == num_of_wins:
                break
            game_counter += 1
        else:
            print("Scissors cuts paper! You lose.")
            computer_points += 1
            if computer_points == num_of_wins:
                break
            game_counter += 1
    elif user_gesture == "scissors":
        if comp_gesture == "paper":
            print("Scissors cuts paper! You win!")
            user_points += 1
            if user_points == num_of_wins:
                break
            game_counter += 1
        else:
            print("Rock smashes scissors! You lose.")
            computer_points += 1
            if computer_points == num_of_wins:
                break
            game_counter += 1
    play_again = input("Continue? (y/n): ")
    if play_again.lower() == "n":
        break
    if game_counter == round_limit:
        break
    print("\n    - - - - - - -\n")
print("\n\n**********************\n")
if game_counter == 1:
    print(f"Game finishes after one round.\nFINAL SCORE: Player {user_points} : "
          f"{computer_points} Computer")
else:
    print(f"Game finishes after {game_counter} rounds.\nFINAL SCORE: Player {user_points} :"
          f" {computer_points} Computer")
```

```python
def how_many_wins():
    """Asks user to enter an integer number and returns 
       tuple of integers in the format (number of rounds, number of wins)"""
    while True:
        try:
            num_wins = int(input("How many rounds to win? "))
        except ValueError:
            print("Please type in a number")
        else:
            if num_wins < 1:
                num_wins = 1
            total_rounds = num_wins * 2 - 1
            return total_rounds, num_wins
```

```python
def get_user_gesture():
    """ask for a gesture, return element of GESTURES"""
    this_gest = "ch405"    # will be assigned user choice
    while this_gest.lower() not in ["r", "s", "p", "rock", "paper", "scissors"]:
        this_gest = input("Enter a choice (r)ock, (p)aper, (s)cissors: ")
        if this_gest.lower() in ["r", "s", "p"]+[GESTURES]:
            break
        print("Please enter a correct choice.")
    if this_gest == "r":
        this_gest = "rock"
    elif this_gest == "p":
        this_gest = "paper"
    elif this_gest == "s":
        this_gest = "scissors"
    return this_gest.lower()
```
